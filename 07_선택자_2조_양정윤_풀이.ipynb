# 문제 2 

선택자를 이용해서 부모 태그의 클래스 태그속성 값이 lecs 인 모든 li 태그를 아래와 같이 출력하시오.

***출력결과***
```
<li>Java 초고수 되기</li>
<li>파이썬 기초 프로그래밍</li>

<li>안드로이드 블루투스 프로그래밍</li>
```

````python
from bs4 import BeautifulSoup

html = """
<html><body>
<div id="main">
  <h1>기초강의목록</h1>
  <ul class="lecs">
    <li>Java 초고수 되기</li>
    <li>파이썬 기초 프로그래밍</li>
  </ul>
</div>
<div id="adv">
  <h1>심화강의목록</h1>
  <ul class="lecs">
    <li>파이썬 머신러닝 프로그래밍</li>
    <li>안드로이드 블루투스 프로그래밍</li>
  </ul>
</div>
</body></html>
"""
soup = BeautifulSoup(html, 'lxml')
for i in soup.select(".lecs > li"):
    print(i)
````

# 문제6
아래 사이트에서 시가총액 Top10 회사들의 이름을 아래와 같이 출력하시오.

http://finance.naver.com/sise/

***출력결과***
```
===시가총액 Top10===
1위: 삼성전자
2위: LG에너지솔루션
3위: SK하이닉스
4위: 삼성바이오로직스
5위: LG화학
6위: 삼성SDI
7위: 삼성전자우
8위: 현대차
9위: 기아
10위: NAVER
```

````python
from bs4 import BeautifulSoup
import requests as rq

url = "http://finance.naver.com/sise/"
html = rq.get(url).text

soup = BeautifulSoup(html, "lxml")
tags = soup.select("#siselist_tab_7 a")

print("===시가총액 Top10===")

for i, v in enumerate(tags):
    print(f"{i+1}위: {v.text}")
````
